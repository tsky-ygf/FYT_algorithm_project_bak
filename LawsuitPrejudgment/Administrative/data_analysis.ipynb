```python
import  pandas as pd
from pprint import pprint
import re
```

```python
data = pd.read_csv("data/DocData/xzcf.csv",low_memory=False)
```

```python
data.head()
```

```python
for index,row in data.iterrows():
    pprint(row.to_dict())
    if index > 10:
        break
```

```python
section_list = data['section'].values
```

```python
agencies_list = []
for section in section_list:
    # print(section)
    try:
        local_ori = re.search("【.*】",section).group()
    except:
        print(section)
        continue
    local = local_ori.replace('【','').replace('】','')
    agencies = section.replace(local_ori,'').replace(local,'')
    # print(local)
    # print(agencies)
    agencies_list.append(agencies)
    # break
```

```python
agencies_list = list(set(agencies_list))
```
